### Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Species'] = le.fit_transform(df['Species'])
df.head()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
proc_df = scaler.fit_transform(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] = proc_df
df.head()
```

```python
## Correlation map 
corr = df.corr()
sns.heatmap(corr, annot=True, cmap = 'coolwarm')
```

**Inference:**   

    1. Species is highly correlated with Petal Width, Petal Length and Sepal Length
    2. Petal Width, Sepal Length and Petal Length are correlated with each other
    3. Sepal Width is not correlated with any feature.
    

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']], df['Species'], test_size=0.25, random_state=0)
x_train.shape
```

### Logistic Regression

```python
## Training
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
```

```python
## Prediction
predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)

print("Accuracy on test data: ", score*100,"%")


print("Precision :", precision_score(y_test, predictions, average='macro'))
print("Recall :", recall_score(y_test, predictions, average='macro'))
print("F1 Score :", f1_score(y_test, predictions, average='macro'))
```

```python
## Confusion Matrix
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predictions)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

**Inference:**

    One instance of class Virginica was misclassified as versicolor

```python
## Decision Boundary
from mlxtend.plotting import plot_decision_regions
value=0
width=1

plot_decision_regions(x_train.values, y_train.values, logisticRegr, feature_index=[2,3], filler_feature_values={0: value, 1:value},
                      filler_feature_ranges={0: width, 1: width}, legend=3)
plt.ylabel('Petal Width (standardized)')
plt.xlabel('Petal Length (standardized)')
plt.title("Decision Boundary produced by Logistic Regression Model", size = 14)
```

### Naive Bayes

```python
## Training
from sklearn.naive_bayes import GaussianNB
GNBclf = GaussianNB()
model = GNBclf.fit(x_train, y_train)
```

```python
## Prediction
preds = GNBclf.predict(x_test)
score = GNBclf.score(x_test, y_test)
print("Accuracy on test data: ", score*100,"%")
print("Precision :", precision_score(y_test, preds, average='macro'))
print("Recall :", recall_score(y_test, preds, average='macro'))
print("F1 Score :", f1_score(y_test, preds, average='macro'))
```

```python
## Confusion Matrix
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, preds)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

```python
## Decision Boundary
plot_decision_regions(x_train.values, y_train.values, GNBclf, feature_index=[2,3], filler_feature_values={0: value, 1:value},
                      filler_feature_ranges={0: width, 1: width}, legend=3)
plt.ylabel('Petal Width (standardized)')
plt.xlabel('Petal Length (standardized)')
plt.title("Decision Boundary produced by Naive Bayes Model", size = 14)
```

### KNN

```python
## Training
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
## Prediction
preds = knn.predict(x_test)
score = knn.score(x_test, y_test)
print("Accuracy on test data: ", score*100,"%")
print("Precision :", precision_score(y_test, preds, average='macro'))
print("Recall :", recall_score(y_test, preds, average='macro'))
print("F1 Score :", f1_score(y_test, preds, average='macro'))
```

```python
x=[]
y=[]
for i in range(3,40):
    knn_temp = KNeighborsClassifier(n_neighbors=i)
    knn_temp.fit(x_train, y_train)
    preds = knn_temp.predict(x_test)
    score = knn_temp.score(x_test, y_test)
    x.append(i)
    y.append(score)


plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.ylabel('Accuracy')
plt.xlabel('k value')
plt.title("Choosing K");
```

```python
## Confusion Matrix
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, preds)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

```python
## Decision boundary
plot_decision_regions(x_train.values, y_train.values, knn, feature_index=[2,3], filler_feature_values={0: value, 1:value},
                      filler_feature_ranges={0: width, 1: width}, legend=3)

plt.ylabel('Petal Width (standardized)')
plt.xlabel('Petal Length (standardized)')
plt.title("Decision Boundary produced by KNN", size = 14)
```
