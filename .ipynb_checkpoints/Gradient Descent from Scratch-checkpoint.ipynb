```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
ct = ColumnTransformer(
     transformers=[("onehotencoding", OneHotEncoder(), ["Region"]),
     ("labelEncoding", OrdinalEncoder(), ["Own",'Student', 'Married']),
      ],
     remainder='passthrough')

#("scaler", StandardScaler(),["Limit", "Income", "Rating"])
```

```python
df = ct.fit_transform(data)
df
```

```python
pd.DataFrame(df)
```

```python
def prediction(X,w):
    return np.dot(X,w).reshape(-1,1)
```

```python
def Losss(error):
    return np.squeeze(np.dot(error.T,error), axis=0)
```

```python
def delLoss(X, error):
    dloss = np.dot(X.T,error)
    return dloss
```

```python
def GradientDescent(X, y, lr, iterations):
    m, dim=X.shape
    w = np.zeros((dim,1))
    loss = []
    for i in range(iterations):
        pred = prediction(X,w)
        error =y-pred 
        
        l= Losss(error)/(2*m)
        loss.append(l)
        
        gi = 2*delLoss(X,error)/m
        w -= lr*gi
        
    return w, loss
```

```python
X=df[:,0:-2]
X=np.hstack((np.ones(X.shape[0]).reshape(-1,1),X))
print(X.shape)

y=np.expand_dims(df[:,-1],1)
print(y.shape)
```

```python
interations = 150
W, Loss = GradientDescent(X, y, 0.000000001, interations)
print(W)
```

```python
plt.plot(range(interations),Loss )
```
